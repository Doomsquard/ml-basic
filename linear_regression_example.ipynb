```python
import pandas as pd # lib for reading docs
import numpy as np # lib for math`s operations
import matplotlib.pyplot as plt # lib for graphs
from sklearn import linear_model # lib for ml

# add medium data to empty Age datas
pd.options.mode.chained_assignment = None # disable notif

# for linear regressions: Y = aX + b

df = pd.read_excel('./stat_docs/price.xlsx') # reading from file

plt.scatter(df.area, df.price, color = "red", marker = "^") # create scatter graph by params
plt.xlabel("Площадь (кв. м.)")
plt.ylabel("Цена (млн. руб.)")

# model training

reg = linear_model.LinearRegression() # create model
reg.fit(df[['area']], df.price) # training model ([[]] - for data, . - for answers)

# predictions
reg.predict([[120]]) # predict for 120 meters

reg.predict(df[['area']]) # predict for training values


reg.coef_ # a value from formula

reg.intercept_ # b value from formula

plt.plot(df.area, reg.predict(df[["area"]])) # create line by predict vals

```

```python
# predict price without answers

pred = pd.read_excel('./stat_docs/predict_price.xlsx')

pred.head() # return only head values

p = reg.predict(pred) # predict for new values
```

```python
# create new column in file with prices

pred['prediction prices'] = p


pred.to_excel('predicted_docs/prices.xlsx', index=False) # create slsx in excel
```

```python
## predict for gas values gdprussia

gdp = pd.read_excel('./stat_docs/gdprussia.xlsx')

gdp.head()
```
